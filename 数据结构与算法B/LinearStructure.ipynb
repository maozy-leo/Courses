# 一、线性结构

线性结构是一种有序数据项的集合，直观而言，数据项之间像一条线一样被前后串接起来：
- 除首尾数据项之外，每个数据项都有唯一的前驱与后继；
- 新数据项通常加到首端或尾端。

对于线性结构而言，不同数据项之间最重要的是先后的次序关系，而不是绝对位置关系。

下面依次介绍栈、队列、双端队列与线性表四种 ADT，并给出基于链表的标准 Python 实现。

### 统一结点定义

下面的实现都复用同一个 `Node` 类。`value` 存放数据项，`next` 指向后继结点，`prev` 指向前驱结点。栈、队列和线性表只依赖 `next` 也能完成操作；双端队列需要同时维护 `prev` 和 `next`。

```python
class Node():
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None
```

## 栈

### ADT 定义
栈（stack）只允许在同一端插入和删除元素，遵循后进先出（LIFO）。
基本操作通常包括：`push`、`pop`、`peek`、`isEmpty`、`size`。

### 链式结构
用单链表实现时，把表头当作栈顶。`push` 和 `pop` 都只操作表头，因此时间复杂度都是 O(1)。

```python
class LinkedStack():
    def __init__(self):
        self.top = None
        self._size = 0

    def push(self, value):
        node = Node(value)
        node.next = self.top
        if self.top is not None:
            self.top.prev = node
        self.top = node
        self._size += 1

    def pop(self):
        if self._size == 0:
            raise IndexError("Pop from empty stack!")
        node = self.top
        self.top = node.next
        if self.top is not None:
            self.top.prev = None
        self._size -= 1
        return node.value
    
    def peek(self):
        if self.top is None:
            raise IndexError("Stack is empty!")
        return self.top.value
    
    def size(self):
        return self._size

    def isEmpty(self):
        return self._size == 0
```

## 队列

### ADT 定义
队列（queue）只允许在一端插入、另一端删除元素，遵循先进先出（FIFO）。
基本操作通常包括：`enqueue`、`dequeue`、`front`、`isEmpty`、`size`。

### 链式结构
用单链表并维护头尾指针：入队发生在尾部，出队发生在头部，因此两者都是 O(1)。

```python
class LinkedQueue():
    def __init__(self):
        self._size = 0
        self._front = None
        self.rear = None

    def enqueue(self, value):
        node = Node(value)
        if self._size == 0:
            self._front = node
            self.rear = node
        else:
            self.rear.next = node
            node.prev = self.rear
            self.rear = node
        self._size += 1

    def dequeue(self):
        if self._size == 0:
            raise IndexError("Dequeue from an empty queue!")
        node = self._front
        self._front = node.next
        if self._front is None:
            self.rear = None
        else:
            self._front.prev = None
        self._size -= 1
        return node.value

    def front(self):
        if self._front is None:
            raise IndexError("Queue is empty!")
        return self._front.value

    def size(self):
        return self._size

    def isEmpty(self):
        return self._size == 0
```

## 双端队列

### ADT 定义
双端队列（deque）允许在两端插入和删除元素。
基本操作通常包括：`addFront`、`addRear`、`removeFront`、`removeRear`、`peekFront`、`peekRear`、`isEmpty`、`size`。

### 链式结构
用双向链表最合适，因为两端操作都只需要改动少量指针。

```python
class LinkedDeque:
    def __init__(self, iterable=None):
        self._size = 0
        self.front = None
        self.rear = None
        if iterable is not None:
            for item in iterable:
                self.addRear(item)

    def addFront(self, item):
        node = Node(item)
        if self._size == 0:
            self.front = node
            self.rear = node
        else:
            node.next = self.front
            self.front.prev = node
            self.front = node
        self._size += 1

    def addRear(self, item):
        node = Node(item)
        if self._size == 0:
            self.front = node
            self.rear = node
        else:
            self.rear.next = node
            node.prev = self.rear
            self.rear = node
        self._size += 1
        
    def removeFront(self):
        if self._size == 0:
            raise IndexError("Remove from an empty deque!")
        node = self.front
        self.front = node.next
        if self.front is None:
            self.rear = None
        else:
            self.front.prev = None
        self._size -= 1
        return node.value

    def removeRear(self):
        if self._size == 0:
            raise IndexError("Remove from an empty deque!")
        node = self.rear
        self.rear = node.prev
        if self.rear is None:
            self.front = None
        else:
            self.rear.next = None
        self._size -= 1
        return node.value

    def peekFront(self):
        if self.front is None:
            raise IndexError("Deque is empty!")
        return self.front.value

    def peekRear(self):
        if self.rear is None:
            raise IndexError("Deque is empty!")
        return self.rear.value

    def __iter__(self):
        current = self.front
        while current is not None:
            yield current.value
            current = current.next

    def size(self):
        return self._size

    def isEmpty(self):
        return self._size == 0
```

## 线性表

### ADT 定义
线性表（list）是由 n ≥ 0 个元素组成的有序序列。这里实现的是无序线性表，即元素没有按大小排序，但仍按插入后的链式次序存放。
基本操作通常包括：`add(item)`、`search(item)`、`remove(item)`、`isEmpty()`、`size()`。

### 链式结构
无序线性表可以用单链表实现。每个结点保存一个元素和指向下一结点的引用，线性表本身只需要保存头结点 `head` 与元素个数。

若采用头插法，`add` 只需修改新结点和头指针，时间复杂度为 O(1)；`search` 和 `remove` 需要从头开始顺序查找，最坏时间复杂度为 O(n)。删除结点时还要记录当前结点的前驱，才能把前驱的 `next` 越过待删结点。

```python
class UnorderedList:
    def __init__(self):
        self.head = None
        self._size = 0

    def add(self, item):
        node = Node(item)
        node.next = self.head
        if self.head is not None:
            self.head.prev = node
        self.head = node
        self._size += 1

    def remove(self, item):
        current = self.head
        previous = None
        found = False
        while current is not None and not found:
            if current.value == item:
                found = True
            else:
                previous = current
                current = current.next
        if not found:
            raise ValueError("Item not found in the list!")
        if previous is None:
            self.head = current.next
            if self.head is not None:
                self.head.prev = None
        else:
            previous.next = current.next
            if current.next is not None:
                current.next.prev = previous
        self._size -= 1

    def search(self, item):
        current = self.head
        found = False
        while current is not None and not found:
            if current.value == item:
                found = True
            else:
                current = current.next
        return found

    def size(self):
        return self._size

    def isEmpty(self):
        return self._size == 0
```
