# 二分查找
二分查找是一种用于在有序列表中查找特定数据项的搜索算法。在每次搜索中，列表从中间项被一分为二，由于列表是有序的，因此我们只需要对比待查找数据项与列表中间值的大小，就可以知道待查找数据项在前半列表或是后半列表中。以此类推，每次我们都将剩余的列表从中间一分为二，重复这一过程，我们要么在某一步查找到数据项，要么直至列表不可再被分割也没有查找到数据项，此时我们断定列表中没有待查找的数据项。

具体而言，以下几个搜索问题是二分查找的典型应用场景。

## Problem 1
在排好序的数组中查找`target`。注意，如果`target`有重复，普通二分查找只能找到其中一个。


```python
from typing import List, Optional

def find_target(nums: List[int], target: int) -> Optional[int]:
    # 假设数组是从小到大排列的
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return None
```

## Problem 2
寻找满足条件`check(item)=True`的第一个项。其中条件`check`具有以下性质：若待查找列表的某一项满足该条件，那么列表后面的所有项都满足这一条件。也就是说，列表按`check`结果看应当呈现`False, False, ..., True, True, ...`的单调结构。

因此，假设待查找项的角标为k，那么k满足：
- `check(items[k-1])=False`
- `check(items[k])=True`

```python
# 我们首先定义条件check函数
def check_positive(item) -> bool:
    # 假设条件为item为正数
    return item > 0

def find_first(items: List[int]) -> Optional[int]:
    if not items:
        return None

    left, right = 0, len(items) - 1
    
    while left < right:
        mid = left + (right - left) // 2
        if check_positive(items[mid]):
            right = mid
        else:
            left = mid + 1
    
    return left if check_positive(items[left]) else None
```

## Problem 3
寻找满足条件`check(item)=True`的最后一个项。其中条件`check`具有以下性质：若待查找列表的某一项满足该条件，那么列表前面的所有项都满足这一条件。也就是说，列表按`check`结果看应当呈现`True, True, ..., False, False, ...`的单调结构。

因此，假设待查找项的角标为k，那么k满足：
- `check(items[k+1])=False`
- `check(items[k])=True`

```python
def check_negative(item) -> bool:
    # 假设条件为item为负数
    return item < 0

def find_last(items: List[int]) -> Optional[int]:
    if not items:
        return None

    left, right = 0, len(items) - 1

    while left < right:
        mid = left + (right - left + 1) // 2
        if check_negative(items[mid]):
            left = mid
        else:
            right = mid - 1

    return left if check_negative(items[left]) else None
```

## 更多应用场景

二分查找的核心不只是“在有序数组里找某个值”，而是在一个有单调性的搜索空间里找边界。

- 查找插入位置：例如 Python 标准库 `bisect_left` / `bisect_right`，可以确定新元素应插入到哪里，或统计某个值出现了多少次。
- 查找第一个不小于目标值的位置：也叫 lower bound，例如在分数线、时间戳、版本号中找第一个达到要求的位置。
- 查找最后一个不大于目标值的位置：也叫 upper bound 的相关变形，例如在日志中找不晚于某个时间的最后一条记录。
- 二分答案：答案本身不是数组元素，但可以判断某个候选答案是否可行。例如最小化最大负载、求最少天数、求最大可行长度、运输能力问题等。
- 连续区间二分：在实数范围内逼近满足条件的值，例如求平方根、求方程近似解。
- 在旋转有序数组、二维有序矩阵中查找：这些问题不是最基础形式，但仍利用有序性逐步缩小搜索范围。
