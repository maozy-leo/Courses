# 零、前置

## “计算”的数学模型

为了解决现实中的问题，我们常常借助“计算”这一方法。形式化地说，计算过程可以理解为一种**基于有穷观点的能行方法**：它由有限条明确规则构成，并且可以被机械地执行。

一个计算过程通常应满足：

- 由有限数量的明确指令构成；
- 每条指令都可以机械执行；
- 在给定状态下，执行结果是确定的；
- 若算法声称求解某问题，则应在有限步后终止并给出结果。

围绕这种思想，数学家和计算机科学家提出了多种等价的计算模型，包括递归函数、lambda 演算、Post 机与图灵机等。其中图灵机最适合用来讨论“什么问题可被算法解决”。

## 图灵机与停机问题

### 图灵机的数学定义

一个（确定性）图灵机可以定义为七元组 $M=(Q,\Sigma,\Gamma,\delta,q_0,q_{acc},q_{rej})$，其中：

- $Q$ 是有限状态集；
- $\Sigma$ 是输入字母表，且空白符号 $\sqcup \notin \Sigma$；
- $\Gamma$ 是纸带字母表，满足 $\Sigma \subseteq \Gamma$ 且 $\sqcup \in \Gamma$；
- $\delta: (Q\setminus\{q_{acc},q_{rej}\})\times\Gamma \to Q\times\Gamma\times\{L,R,S\}$ 是转移函数；
- $q_0 \in Q$ 是初始状态；
- $q_{acc} \in Q$ 是接受状态；
- $q_{rej} \in Q$ 是拒绝状态，且 $q_{rej}\neq q_{acc}$。

图灵机的配置由当前状态、纸带内容与读写头位置共同决定。若记某个配置为 $C$，则由转移函数 $\delta$ 可定义一步转移关系 $C \vdash C'$。若从初始配置出发，经过有限次转移后到达接受状态，则称该输入被机器接受；若到达拒绝状态，则称该输入被拒绝。

### 停机问题

设 $\langle M,w\rangle$ 表示图灵机 $M$ 与输入串 $w$ 的编码。停机问题定义为语言

$$
\text{HALT} = \{\langle M,w\rangle \mid M\text{ 在输入 }w\text{ 上会停机}\}.
$$

问题是：是否存在一个图灵机 $H$，对任意输入 $\langle M,w\rangle$ 都能正确判断 $M$ 在 $w$ 上是否停机？

### 停机问题的不可判定性

下面用反证法证明不存在这样的通用判定机。

假设存在一个图灵机 $H$，对任意 $\langle M,w\rangle$，若 $M$ 在 $w$ 上停机则 $H$ 接受；若 $M$ 在 $w$ 上不停机则 $H$ 拒绝。

构造一台新图灵机 $D$，其输入为某台图灵机 $M$ 的编码 $\langle M\rangle$，并执行如下过程：

1. 将 $\langle M,\langle M\rangle\rangle$ 输入给 $H$；
2. 若 $H$ 接受，则 $D$ 进入无限循环；
3. 若 $H$ 拒绝，则 $D$ 停机接受。

现在考虑 $D$ 自身的编码 $\langle D\rangle$ 并令输入也为 $\langle D\rangle$。分析 $D(\langle D\rangle)$：

- 若 $H$ 判断 $D(\langle D\rangle)$ 会停机，则按构造 $D$ 反而进入无限循环，因此它不停机；
- 若 $H$ 判断 $D(\langle D\rangle)$ 不会停机，则按构造 $D$ 会停机接受。

这与 $H$ 的正确性假设矛盾，因此这样的判定机不存在。故停机问题不可判定。

这个结论说明：计算理论不仅关心“哪些问题能算”，也关心“哪些问题根本不存在通用算法”。

## 数据结构与算法

**数据结构**研究数据的组织方式。不同组织方式会影响存储、访问、插入、删除、查找和更新效率。常见数据结构包括数组、链表、栈、队列、树、图、堆与散列表等。

**算法**研究解决问题的步骤。一个算法需要说明输入、输出、执行过程、终止条件与正确性。常见思想包括枚举、递归、分治、贪心、动态规划、回溯、搜索、排序和图算法等。

二者通常要结合起来看。合适的数据结构会让算法更自然、更高效；算法需求也会反过来决定应选择怎样的数据结构。

## 算法分析简介

设算法的输入规模为 $n$。算法分析关注的是当 $n$ 增大时，算法所需资源的增长趋势，而不是某台机器上的精确运行秒数。最常见的分析对象是时间复杂度和空间复杂度。

### 大 O 记号

设 $f(n)$ 和 $g(n)$ 为定义在足够大整数上的非负函数。若存在常数 $c>0$ 和 $n_0$，使得对一切 $n\ge n_0$ 都有

$$
0 \le f(n) \le c\,g(n),
$$

则称 $f(n)=O(g(n))$。

它表示 $f$ 的增长至多与 $g$ 同阶，常用于给出复杂度上界。

### 大 $\Omega$ 记号

若存在常数 $c>0$ 和 $n_0$，使得对一切 $n\ge n_0$ 都有

$$
0 \le c\,g(n) \le f(n),
$$

则称 $f(n)=\Omega(g(n))$。

它表示 $f$ 的增长至少与 $g$ 同阶，常用于给出复杂度下界。

### 大 $\Theta$ 记号

若同时有 $f(n)=O(g(n))$ 且 $f(n)=\Omega(g(n))$，则称

$$
f(n)=\Theta(g(n)).
$$

等价地，存在常数 $c_1,c_2>0$ 和 $n_0$，使得对一切 $n\ge n_0$ 都有

$$
c_1 g(n) \le f(n) \le c_2 g(n).
$$

它表示 $f$ 与 $g$ 同阶增长。

### 举例说明

若 $f(n)=3n+10$，则有 $f(n)=O(n)$、$f(n)=\Omega(n)$，因此 $f(n)=\Theta(n)$。

若 $f(n)=n^2+5n+7$，则有 $f(n)=O(n^2)$、$f(n)=\Omega(n^2)$，因此 $f(n)=\Theta(n^2)$。

若 $f(n)=\log n$，则有 $f(n)=O(n)$，但不可能有 $f(n)=\Omega(n)$。

分析复杂度时常用以下原则：

- 顺序执行的代码段，总复杂度取增长最快的一项；
- 嵌套循环的复杂度通常相乘；
- 递归算法常结合递归树、递推式或主定理分析；
- 常数因子和低阶项通常被忽略。

## Python 数据类型的性能

Python 内置类型已经封装了许多常见数据结构。理解它们的典型复杂度，有助于避免不必要的低效操作。

### `list` 常见操作的复杂度

| 操作 | 平均复杂度 | 说明 |
| --- | --- | --- |
| `a[i]` | $O(1)$ | 按下标访问 |
| `a[i] = x` | $O(1)$ | 按下标赋值 |
| `append(x)` | 均摊 $O(1)$ | 尾部追加 |
| `pop()` | 均摊 $O(1)$ | 删除尾部元素 |
| `insert(i, x)` | $O(n)$ | 需要移动元素 |
| `pop(i)` | $O(n)$ | 需要移动元素 |
| `remove(x)` | $O(n)$ | 先查找再删除 |
| `x in a` | $O(n)$ | 线性查找 |
| `index(x)` | $O(n)$ | 线性查找 |
| `count(x)` | $O(n)$ | 遍历计数 |
| `extend(b)` | $O(k)$ | $k$ 为追加的元素个数 |
| `reverse()` | $O(n)$ | 原地翻转 |
| `sort()` | $O(n\log n)$ | Python 的 Timsort |
| 切片 `a[i:j]` | $O(k)$ | $k$ 为切片长度 |
| 迭代遍历 | $O(n)$ | 按元素顺序访问 |

### 哈希对象的复杂度

对于 `dict`，常见操作的平均复杂度为：

| 操作 | 平均复杂度 | 说明 |
| --- | --- | --- |
| `d[k]` | $O(1)$ | 按键取值 |
| `d[k] = v` | $O(1)$ | 插入或更新 |
| `del d[k]` | $O(1)$ | 删除键 |
| `k in d` | $O(1)$ | 键成员判断 |
| `get(k)` | $O(1)$ | 安全取值 |
| `keys()` / `values()` / `items()` | $O(1)$ 生成视图 | 视图本身的创建是常数时间 |
| 遍历字典 | $O(n)$ | 与元素个数成正比 |

字典依赖哈希表。对键的访问先计算哈希值，再在对应桶中定位，因此平均情况下查找、插入和删除都很快。`set` 的情形与 `dict` 类似，只是它存储的是元素本身而不是键值对。

### 什么时候会违反上述复杂度

上面的复杂度通常是平均意义或均摊意义上的结论，以下情况可能使实际表现偏离：

- **哈希冲突严重**：`dict` 或 `set` 中大量键落入同一桶，查找可能退化；
- **恶意输入**：如果键的构造刻意制造冲突，哈希表的性能会明显下降；
- **动态扩容**：`list.append()`、`dict` 插入等操作在触发扩容时会有一次较大的搬移或重哈希开销；
- **元素类型本身开销大**：即使主操作是 $O(1)$，若键的哈希计算或比较本身很慢，整体时间也会变大；
- **最坏情况分析**：复杂度表中的平均值不等于最坏值，某些操作在理论上可能退化到更高阶。

因此，复杂度表描述的是典型行为，实际性能还要结合数据规模、数据分布和具体实现一起看。
