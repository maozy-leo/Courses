# 二叉堆

优先队列是一种特殊的队列数据结构，与普通队列的“先进先出(FIFO)”不同，优先队列中的元素会依据优先级（元素的大小，或是其他全序关系）：对于最大优先队列，优先级最高的元素先出队；最小优先队列则相反。在以下介绍中，我们主要考虑最小优先队列，相关的算法及实现对于最大优先队列几乎是一样的。

与队列一致，优先队列需要支持`push`、`pop`、`top`、`is_empty`、`size`等操作。

为了实现一个高效的优先队列数据结构，我们首先介绍完全二叉树的列表实现。

## 完全二叉树的列表实现
不同于一般的树结构，完全二叉树由于其结构的特殊性，我们简单地用列表进行储存与操作。

对于一棵给定的完全二叉树，我们对其进行层次遍历（BFS），得到一个连续的序列，我们将这个序列存储在列表`lst`中就得到了完全二叉树的列表实现。

我们首先对这个列表进行较为精细的研究。对于完全二叉树第 $k$ 层的第 $i$ 个元素（注：k与i均从0开始计算），它在列表中的序号为
$$
2^0 + 2^1 + \dots + 2^{k-1} + i = 2^{k} + i - 1
$$
这个节点的左/右子节点（如有）分别为完全二叉树第 $k+1$ 层的第 $2i$ 与第 $2i+1$ 个元素。它们在列表中的序号分别为 $2^{k+1}+2i-1$ 与 $2^{k+1}+2i$。

因此，经过简单的计算，我们可以得到结论：对于完全二叉树的元素`lst[k]`，它的父节点为`lst[(k-1)//2]`，它的左右子节点分别为`lst[2*k+1]`与`lst[2*k+2]`。

## 二叉堆
堆是一种特殊的完全二叉树，分为最大堆与最小堆。最大堆的每个节点的值不小于它的子节点的值。最小堆则相反。

优先队列可以通过堆实现。各个操作的时间复杂度分别为：
- `push`: $O(\log N)$;
- `pop`: $O(\log N)$;
- `top`: $O(1)$;
- 从长度为$N$的数组建堆： $O(N)$。



```python
class BinaryHeap:  # 在这里我们实现的是最小堆，最大堆几乎是完全一样的
    def __init__(self, nums: list[int] | None = None):
        self._heap = self.buildHeap(nums or [])
        self.size = len(self._heap)

    @staticmethod
    # 对堆中的每个非叶子节点倒序执行下降操作
    def buildHeap(nums: list[int]) -> list[int]:
        heap = list(nums)
        for idx in range(len(heap) // 2 - 1, -1, -1):
            BinaryHeap._shiftDown(heap, idx, len(heap))
        return heap

    def top(self):
        if self.is_empty():
            raise IndexError("top from empty heap")
        return self._heap[0]

    def push(self, value):
        self._heap.append(value)
        self.size += 1
        self._shiftUp(self._heap, self.size - 1)

    def pop(self):
        if self.is_empty():
            raise IndexError("pop from empty heap")

        self._heap[0], self._heap[self.size - 1] = self._heap[self.size - 1], self._heap[0]
        value = self._heap.pop()
        self.size -= 1

        if not self.is_empty():
            self._shiftDown(self._heap, 0, self.size)
        return value

    def is_empty(self):
        return self.size == 0

    def __len__(self):
        return self.size

    @staticmethod
    def _shiftUp(nums: list[int], idx: int) -> None:
        while idx > 0:
            parent = (idx - 1) // 2
            if nums[parent] <= nums[idx]:
                break
            nums[parent], nums[idx] = nums[idx], nums[parent]
            idx = parent

    @staticmethod
    def _shiftDown(nums: list[int], idx: int, size: int) -> None:
        while True:
            left = idx * 2 + 1
            right = idx * 2 + 2
            smallest = idx

            if left < size and nums[left] < nums[smallest]:
                smallest = left
            if right < size and nums[right] < nums[smallest]:
                smallest = right

            if smallest == idx:
                break

            nums[idx], nums[smallest] = nums[smallest], nums[idx]
            idx = smallest
```
